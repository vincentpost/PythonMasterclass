## Pandas

Some data

```python
height = [150, 155, 164, 174, 183, 189, 192, 192]
age = [11, 12, 13, 14, 15, 16, 17, 18]
```

Things you can do with this data:
1. get the height at the age of 13
2. get the mean height between the ages of 11 and 18
3. plot the height vs the age
4. convert to inches

```python
# 1. get the height at the age of 13
i = age.index(13)
height[i]
```

```python
# 2. get the mean height between the ages of 11 and 18
import numpy as np
np.mean(height)
```

```python
# 3. plot the height vs the age
import matplotlib.pyplot as plt
plt.plot(age, height)
```

```python
# 4. convert to inches
height_inch = []
for h in height:
    h_inch = h * 0.3937
    height_inch.append(h_inch)
height_inch
```

## Pandas Series

Doing the same 3 things using pandas.

```python
import pandas as pd

s = pd.Series([150, 155, 164, 174, 183, 189, 192, 192], 
              [11, 12, 13, 14, 15, 16, 17, 18])
s
```

```python
# 1. get the height at the age of 13
s.loc[13]
```

```python
# 2. get the mean height between the ages of 11 and 18
s.mean()
```

```python
# 3. plot the height vs the age
s.plot()
```

```python
# 4. convert to inches
s_inch = s * 0.3937
s_inch
```

## Pandas DataFrame

Add dimension.


```python
df = pd.DataFrame(index=range(11,19,1),
                  data= {'height': [150, 155, 164, 174, 183, 189, 192, 192],
                         'weight': [42, 46, 50, 53, 55, 58, 62, 65]})
df
```

```python
# plot data
df.plot()
```

```python
# plot using a secondary y axis
df.plot(secondary_y=["weight"])
```

```python
# creating a new column calculated from the other two
df['BMI'] = df['weight'] / ((df['height']/100) **2)
df
```

## Data types

```python

df = pd.DataFrame(index=['2010-1-1', '2010-5-25', '2010-7-23', '2010-8-31'],
                  data= {'groundwater table': [153, 133, 131, 150],
                         'surface water level': [222, 153, 120, 144]})
df
```

```python
df.plot()
```

```python
# show the datatype of the index
df.index
```

```python
# convert the index to a datetime index
df.index= pd.to_datetime(df.index)
```

```python
# plot the data using the date time axis.
df.plot(marker='o')
```
