## Pandas

Some data

This is a simple example of how to use panadas 

```python
height = [150, 155, 164, 174, 183, 189, 192, 192]
age = [11, 12, 13, 14, 15, 16, 17, 18]
```

Things you can do with this data:
1. get the height at the age of 13
2. get the mean height between the ages of 11 and 18
3. plot the height vs the age
4. convert to inches

```python
# 1. get the height at the age of 13
# this code gets you the position of the element with age 13
i = age.index(13)
print( r'The height is :' + str(height[i]) )
```

```python
# 2. get the mean height between the ages of 11 and 18
import numpy as np
np.mean(height)
```

```python
# 3. plot the height vs the age
import matplotlib.pyplot as plt
plt.plot(age, height)
```

```python
# 4. convert to inches
height_inch = []
for h in height:
    h_inch = h * 0.3937
    height_inch.append(h_inch)
height_inch
```

## Pandas Series

Doing the same 3 things using pandas.

```python
import pandas as pd

s = pd.Series([150, 155, 164, 174, 183, 189, 192, 192], 
              [11, 12, 13, 14, 15, 16, 17, 18])
s
```

```python
# 1. get the height at the age of 13
s.loc[13]
```

```python
# 2. get the mean height between the ages of 11 and 18
s.mean()
```

```python
# 3. plot the height vs the age
s.plot()
```

```python
# 4. convert to inches
s_inch = s * 0.3937
s_inch
```

## Pandas DataFrame

Add dimension.


```python
import pandas as pd

df = pd.DataFrame(index=range(11,19,1),
                  data= {'height': [150, 155, 164, 174, 183, 189, 192, 192],
                         'weight': [42, 46, 50, 53, 55, 58, 62, 65]})

# With this command you can check the data type of each column:

df.dtypes
```

Here you convert the dates into a date time object 

```python
dates = pd.to_datetime(['2010-1-1', '2010-5-25', '2010-7-23', '2010-8-31'])
print(dates)
```

```python
# plot data
df.plot()
```

```python
# plot with markers 

df.plot(marker='o')
```

```python
# plot using a secondary y axis
df.plot(secondary_y=["weight"])
```

```python
# creating a new column calculated from the other two
df['BMI'] = df['weight'] / ((df['height']/100) **2)
df
```
